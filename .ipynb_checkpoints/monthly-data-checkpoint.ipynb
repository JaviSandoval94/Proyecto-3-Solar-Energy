# Month 

```python
# monthly_data['month_year'] = pd.to_datetime(monthly_data['date']).dt.to_period('M')
monthly_data['month_year'] = monthly_data['date'].dt.floor('d') - pd.offsets.MonthBegin(1)
monthly_data
```

```python
monthly_data = monthly_data[['id', 'production(kWh)', 'specific_prod(kWh/kWp)', 'uvIndex',
       'sunHour', 'cloudcover', 'maxtempC', 'mintempC', 'precipitation', 'month_year']]
monthly_data.head()
```

```python
monthly_data = monthly_data.rename(columns = {'month_year': 'date'})
```

```python
monthly_data.head()
```

## Sum

```python
sum_df = monthly_data.groupby(by=['id', 'date']).sum()
sum_df = sum_df[['production(kWh)', 'specific_prod(kWh/kWp)','precipitation']]
sum_df.head()
```

```python
sum_df.reset_index(level=0, inplace=True)
```

```python
sum_df.reset_index(level=0, inplace=True)
```

```python
sum_df = sum_df[['id', 'date', 'production(kWh)', 'specific_prod(kWh/kWp)',
       'precipitation']]
```

```python
sum_df
```

## Mean 

```python
mean_df = monthly_data.groupby(by=['id', 'date']).mean()
mean_df = mean_df[['uvIndex', 'sunHour', 'cloudcover']]
mean_df.head()
```

```python
mean_df.reset_index(level=0, inplace=True)
```

```python
mean_df.reset_index(level=0, inplace=True)
```

```python
mean_df
```

## Max

```python
max_df = monthly_data.groupby(by=['id', 'date']).max()
max_df = max_df[['maxtempC']]
max_df.head()
```

```python
max_df.reset_index(level=0, inplace=True)
```

```python
max_df.reset_index(level=0, inplace=True)
```

```python
max_df
```

## Min

```python
min_df = monthly_data.groupby(by=['id', 'date']).min()
min_df = min_df[['mintempC']]
min_df.head()
```

```python
min_df.reset_index(level=0, inplace=True)
```

```python
min_df.reset_index(level=0, inplace=True)
```

```python
min_df
```

### Merge Monthly

```python
final_df = pd.merge(min_df, max_df, how='left', on=['date', 'id'])
final_df.head()
```

```python
final_df = pd.merge(final_df, sum_df, how='left', on=['date', 'id'])
final_df.head()
```

```python
final_df = pd.merge(final_df, mean_df, how='left', on=['date', 'id'])
final_df.head()
```
